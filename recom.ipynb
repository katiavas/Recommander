```python
from langchain_core.documents import Document
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

documents = [
    Document(page_content=line, metadata={"source": "tagged_description.txt"})
    for line in raw_documents[0].page_content.split("\n")
    if line.strip()
]
```

How do get access to openai API

```python
'''Build vector database
Create document embeddings and store them i a document database
Use chroma as database , use the from documents method from chroma, pass the documents and pass an embedding.
We use openai embeddings
 '''

# Create vector database
db_books = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings())
```

```python
#1st example query
query = "A book to teach children about nature"
# take the vector database and perform a similarity search with the query
docs = db_books.similarity_search(query, k = 10)
# Get the recommendation
docs
```

This is just returning recomendations of descriptions to use but we want recommendation of book titles and authors. So we need a way to use these recommendations to filter the dataframe containing all of our books.  So this is where we use the ISBNS.

```python
# Get the ISBN
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
# Since before we got one recommendation for our query, lets put it a function to get  multiple reccommendations .
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
